```python
import numpy as np
```

```python
def unit_step_fun(x):
    return np.where(x>0, 1, 0)

def relu(x):
    return np.maximum(0,x)

def leakyrelu(x):
    return max(0.001*x, x)

def sigmoid(x):
    return 1/(1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10, 10, 1000) 
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x)
```

```python
plt.plot(x, sigmoid(x))
```

```python
plt.plot(x, tanh(x) )
```

```python
plt.plot(x, relu(x) )
```
