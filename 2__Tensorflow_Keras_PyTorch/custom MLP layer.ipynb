```python
import torch
import torch.nn as nn
```

```python
class LinearReLU(nn.Module):
    def __init__(self, input_size, output_size):
        super(LinearReLU, self).__init__()
        self.linear = nn.Linear(input_size, output_size)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.linear(x)
        x = self.relu(x)
        return x
```

```python
model = nn.Sequential(
    nn.Flatten(),
    LinearReLU(784, 256),
    LinearReLU(256, 128),
    nn.Linear(128, 10)
)
```
