```python
import tensorflow as tf

a = tf.constant(5)
b = tf.constant([1,2,3])
c = tf.constant([[1,2], [3,4]])

print(a)
print(b)
print(c)
```

```python
x = tf.constant([[1, 2, 3], [4, 5, 6]])

print("Tensor:", x)
print("Rank:", tf.rank(x))        
print("Shape:", x.shape)        
print("Dtype:", x.dtype)
```

### Special Tensors

```python
zeros = tf.zeros((2,3))
ones = tf.ones((3,3))
random_tensor = tf.random.normal((2,2), mean=0, stddev=1)

print(zeros)
print(ones)
print(random_tensor)
```

### Operations

#### Element wise operations

```python
x = tf.constant([2, 4, 6])
y = tf.constant([1, 1, 1])

print('Add:', x+y)
print('Multiply:', x*y)
print('Square:', tf.square(x))
```

#### Matrix Multiplication

```python
a = tf.constant([[1, 2, 3]])
b = tf.constant([[4], [5], [6]])

print(a.shape, b.shape)

c = tf.matmul(a,b)
print('Matrix Multiplication Result:', c)
```

#### Broadcasting

```python
a = tf.constant([[1, 2, 3]])
b = tf.constant(2)

print(a+b)
```

```python
a = tf.constant([[1, 2, 3],[4, 5, 6]])
b = tf.constant([10, 20, 30])

print(a+b)
```

```python
a = tf.constant([[1],
                 [2],
                 [3]])
b = tf.constant([10, 20, 30])

print(a + b)
```

```python
a = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]]) 
b = tf.constant([10, 20, 30])  

# print(a+b) --fails
```

Broadcasting FAILS when any dimension pair (right → left):
- are different
- and not equal to 1
- and not missing

#### Indexing and Slicing

```python
x = tf.constant([[10, 20, 30],
                 [40, 50, 60]])

print("First row:", x[0])
print("Element (1,1):", x[1, 1])
print("All rows, first two columns:\n", x[:, :2])
```

### Example: Graph Created Automatically Using tf.function

```python
import tensorflow as tf

@tf.function
def compute(x):
    return tf.square(x) + tf.sin(x)

result = compute(tf.constant(3.0))
print(result)
```

### Graph in a Keras Model (TF2)

Even though you write Keras code normally, training uses a graph internally.

```python
from tensorflow.keras import layers, Model

inputs = layers.Input(shape=(4,))
x = layers.Dense(8, activation="relu")(inputs)
outputs = layers.Dense(3)(x)

model = Model(inputs, outputs)
model.compile(optimizer="adam", loss="mse")

model.fit(tf.random.normal((10, 4)), tf.random.normal((10, 3)), epochs=1)
```
