## sequence of number

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Input, LSTM
```

```python
X = np.array([
    [1, 2, 3],
    [2, 3, 4],
    [3, 4, 5]
])

y = np.array([4, 5, 6])
```

```python
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
model = Sequential([
    Input(shape=(3,1)),
    SimpleRNN(10, activation='tanh'),
    Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse'
)

model.summary()
```

```python
model.fit(X, y, epochs=300, verbose=0)
```

```python
test_input = np.array([[4, 5, 6]])
test_input = test_input.reshape((1, 3, 1))

prediction = model.predict(test_input)
print(prediction)
```

## LSTM

```python
model = Sequential([
    Input(shape=(3, 1)),
    LSTM(50),
    Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse'
)

model.summary()
```

```python
X = np.array([
    [1, 2, 3],
    [2, 3, 4],
    [3, 4, 5]
])

y = np.array([4, 5, 6])
```

```python
X = X / 100.0
y = y / 100.0
```

```python
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
model.fit(X, y, epochs=300, verbose=0)
```

```python
test_input = np.array([[4, 5, 6]]) / 100.0
test_input = test_input.reshape((1, 3, 1))

prediction = model.predict(test_input)
print("Predicted next number:", prediction[0][0] * 100)
```
