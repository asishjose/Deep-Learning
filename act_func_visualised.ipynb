```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum(axis=0)
```

```python
x = np.linspace(-10, 10, 1000) 
```

```python
fig, axs = plt.subplots(3, 2, figsize=(10, 10))
axs[0, 0].plot(x, sigmoid(x), color='blue')
axs[0, 0].set_title('Sigmoid')

axs[0, 1].plot(x, tanh(x), color='orange')
axs[0, 1].set_title('Tanh')

axs[1, 0].plot(x, relu(x), color='green')
axs[1, 0].set_title('ReLU')

axs[1, 1].plot(x, leaky_relu(x), color='red')
axs[1, 1].set_title('Leaky ReLU')

axs[2, 0].plot(x, softmax(x), color='purple')
axs[2, 0].set_title('Softmax')

axs[2, 0].plot(x, softmax(x), color='purple')
axs[2, 0].set_title('Softmax')

plt.suptitle("Activation Functions", fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
