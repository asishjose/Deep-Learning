```python
image = cv2.imread('soccer_practice.jpg', cv2.IMREAD_GRAYSCALE)
template = cv2.imread('ball.PNG', cv2.IMREAD_GRAYSCALE)
```

```python
image = cv2.resize(image,(0,0), fx=0.7, fy=0.7)
template = cv2.resize(template,(0,0), fx=0.7, fy=0.7)
```

```python
plt.imshow(image, cmap='gray')
```

```python
plt.imshow(template, cmap='gray')
```

```python
match = cv2.matchTemplate(image, template, cv2.TM_CCOEFF )
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(match)
```

```python
h, w = template.shape
```

```python
location = maxLoc
```

```python
bottomright = (location[0]+w, location[1]+h)
```

```python
image = cv2.rectangle(image, location, bottomright,(255,255,255), 5)
```

```python
plt.imshow(image, cmap='gray')
```
